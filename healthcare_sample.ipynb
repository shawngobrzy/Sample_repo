```python
#Read all of the characters of the text into a dataframe as single character values.
with open('medquad.csv', 'r', encoding='utf-8') as f:
    text = f.read()
```

```python
#Print out the length of the characters and the first 1000 characters in the dataset.
print("The number of characters: ", len(text))

print("The first 1000 characters in the text:\n",text[:1000])
```

```python
#Find all of the unique characters in the dataset
characters = sorted(list(set(text)))

unique_characters = len(characters)
print("List of Unique Characters:\n", ''.join(characters))
print("\nTotal number of Unique Characters: ",unique_characters)
```

```python
#Start tokenization of characters to convert from string to integer
string_to_int = { ch:i for i,ch in enumerate(characters) }
int_to_string = { i:ch for i,ch in enumerate(characters) }
encode = lambda s: [string_to_int[c] for c in s] #Convert a string to a list of integers
decode = lambda r: ''.join([int_to_string[i] for i in r]) #Convert a list of integers to a string

#Test the encoder and decoder
print(encode("glaucoma is"))
print(decode(encode("glaucoma is")))
```

```python
#Make the training and testing splits of 80/20 as well as a number to split the predictive blocks (Maximum possible length of predictions) into.
split = int(0.8 * len(sample_data)) 
train_data = sample_data[:split]
test_data = sample_data[split:]

block_size = 7 
train_data[:block_size+1]
```

```python
#Train the data to use the current information to correctly predict the target in the block.
x = train_data[:block_size]
y = train_data[1:block_size+1]
for t in range(block_size):
    current_info = x[:t+1]
    target = y[t]
    print(f"When input is {current_info} the target is: {target}")
```
